# **Solution**

### The project involves:
- Loading the NYC public school SAT performance dataset into a Pandas DataFrame.
- Conducting various analyses to answer key questions about NYC public school SAT performance, such as identifying top-performing schools in math, ranking the top ten schools based on overall average SAT scores, and determining the borough with the highest standard deviation in SAT scores.
- Summarizing the findings and drawing conclusions about the performance of NYC public schools based on SAT scores.

```python
# Import pandas package
import pandas as pd

# Read the data from csv
schools = pd.read_csv("schools.csv")

# Create a df that contains the required columns by selecting using column indices satisfying the required condition
best_math_schools = schools[schools["average_math"] >= 0.8 * 800][["school_name", "average_math"]].sort_values("average_math", ascending=False).reset_index(drop=True)

# Print df to validate code
print(best_math_schools)
```

```python
# Import the pandas package
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Create a new column in `schools` df that sums the average scores across sections for each school
schools["total_SAT"] = schools.loc[:, "average_math":"average_writing"].sum(axis=1)

#  Create a new df containing the required columns
top_10_schools = schools[["school_name", "total_SAT"]].sort_values("total_SAT", ascending=False).head(10).reset_index(drop=True)

# Print df to validate code
print(top_10_schools)
```

```python
# Import the pandas package
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Create a new column in `schools` df that sums the average scores across sections for each school
schools["total_SAT"] = schools.loc[:, "average_math":"average_writing"].sum(axis=1)

# Group the data by borough and calculate aggregates
schools_by_borough = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)

# Rename the columns
schools_by_borough.rename(columns={'count': 'num_schools', 'mean': 'average_SAT', 'std': 'std_SAT'}, inplace=True)

# Select the column with the largest standard deviation
largest_std_dev = schools_by_borough[schools_by_borough["std_SAT"] == schools_by_borough["std_SAT"].max()]

# Print df to validate code
print(large_std_dev)
```
